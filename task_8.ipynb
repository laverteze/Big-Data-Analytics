What percentage of employees who have a Talk Duration of more than 2 minutes (120 seconds)?

```python
tmp2 = df['Talk Duration (AVG)'].str.split(':')

df['hours_talk'] = ((tmp2.str[0]).astype(int))*360

df['minutes_talk'] = ((tmp2.str[1]).astype(int))*60

df['second_talk'] = (tmp2.str[2]).astype(int)
df['avg_talk_duration'] = ( df['hours_talk'] + df['minutes_talk'] + df['second_talk'] )
df3 = df
df3 = df3.drop(columns=['hours','minutes','second','date_int'])
df5 = df3
percent = len(df5[df5['avg_talk_duration'] > 120]) / len(df5['avg_talk_duration'])
percent  = round(percent * 100, 2)
print(percent)
```

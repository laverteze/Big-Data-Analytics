**Data Description:**

- __ID__ - Уникальный номер атлета (Athlete's unique number)
- __Name__ - Имя атлета
- __Sex__ - Пол 
- __Age__ - Возраст 
- __Height__ - Рост (см / cm) 
- __Weight__ - Вес (кг / kg) 
- __Team__ - Название команды 
- __NOC__ - Код Нац.олимп. коммитета (National Olympic Committee 3-letter code)
- __Games__ - Год и сезон проведения игр (Year and season of games)
- __Year__ - Год проведения игр
- __Season__ -сезон проведения игр
- __City__ - Город проведения игр
- __Sport__ - Вид спорта
- __Event__ - Событие
- __Medal__ - Медаль (Gold, Silver, Bronze, или NA)

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('olympic_games.csv', sep = ',') # data import df
df
```

```python
df.info()
```

```python
df
df['Age'].mean(), df['Age'].median() # mean and median age output among all sportsmen
```

```python
df[df['Age']==26] # to round up age of sportsmen
```

```python
df.groupby('Medal')['Age'].describe() # statistics of medals by age
```

```python
df['Weight'].describe() # taking info about weight of the sportsmen
```

```python
df['Weight'].plot(kind = 'hist', color = 'green') # making a histogram of sportsmens' weight
```

```python
df['Height'].describe() # taking info about height of the sportsmen
```

```python
df['Height'].plot(kind = 'hist', color = 'red') # making a histogram of sportsmens' height
```

```python
df.sort_values(['Weight','Name'], ascending=[False,True]) 
```

```python
df.sort_values(['Height','Name'], ascending=[False,True]) 
```

```python
sns.scatterplot(data=df, x="Weight", y="Height", color="b"); # making a scatterplot with weight/height info
```

```python
df.loc[(df['Sex']=='M')& (df['Medal']=='Gold')] # selection of men with golden medals
```

```python
df.loc[(df['Sex']=='F')& (df['Medal']=='Gold')] # selection of women with golden medals
```

```python
df['Medal'].isna().sum() # taking all number of sportsmen with medals
```

```python
df['Medal'].value_counts() # dividing all medalists within their type of medal
```

```python
df.groupby('Medal')['Sex'].value_counts() # dividing all medalists within their type of medal and sex
```

```python
df.groupby(['Year','Sex'])['Age'].agg(['mean']) # grouping statistics of age by year and sex
```

```python
df.groupby(['Year','Medal'])['Age'].agg(['mean', 'median']) # rounding up statistics of age and sex and getting mean and median age
```

```python
df.corr() # data correlations
```

```python
sns.heatmap (df.corr()); # heatmap by correlations
plt.rcParams ['figure.figsize'] = [12,10] # defining a plot size
```
